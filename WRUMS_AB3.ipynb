### Aufgabe 1e) Mosaikplot (Art = Rechteckbreite; Ort bedingt auf Art = Rechteckhöhe)

```python
import plotly.graph_objects as go
import pandas as pd

# our Data of our Table
table = pd.DataFrame({
    "Baum": [3, 0, 1, 6],
    "Teich": [1, 5, 1, 1],
    "Wiese": [2, 1, 1, 2]
}, index = ["Amsel", "Ente", "Raben", "Spatz"])

# total number of observations
N = table.values.sum()

fig = go.Figure()

x_start = 0  # first x position

for tier in table.index:
    # width = proportion of total observations for animals
    width = table.loc[tier].sum() / N
    
    y_start = 0  # y position reset for each animal type
    
    for ort in table.columns:
        height = table.loc[tier, ort] / table.loc[tier].sum()
        
        # add rectangle
        fig.add_shape(
            type="rect",
            x0=x_start, x1=x_start + width,
            y0=y_start, y1=y_start + height,
            line=dict(color="black"),
            fillcolor="lightgrey",
            layer = "below"
        )
        
        # add text
        fig.add_trace(go.Scatter(
            x=[x_start + width/2],
            y=[y_start + height/2],
            text=f"{tier}<br>{ort}<br>{table.loc[tier, ort]}",
            mode="text",
            textposition="middle center",
            textfont=dict(color="black", size = 14),
            showlegend=False
        ))
        
        y_start += height  # put y up for next location
    
    x_start += width  # put x right for next animal type

fig.update_layout(
    title="Mosaikplot: Art × Ort",
    xaxis=dict(showgrid=False, zeroline=False, title="Tierart"),
    yaxis=dict(showgrid=False, zeroline=False, title="Ort"),
    width=800,
    height=600
)

fig.show()
```
